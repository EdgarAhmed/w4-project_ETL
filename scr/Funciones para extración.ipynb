```python
def extraer_our_world_in_data(url):
    
    driver = webdriver.Chrome()
    driver.get (url)

    time.sleep(2)

    # cookies

    aceptar = driver.find_element(By.XPATH, '/html/body/div[1]/div/div/div/div[2]/button[2]')

    aceptar.click()

    time.sleep(2)

    #Data

    clic_data = driver.find_element (By.XPATH, '/html/body/main/figure/div/footer/div/div[2]/ul/li[1]/div/button')

    clic_data.click()

    time.sleep(2)

    #descarga de archivos

    descarga1 = clic_data = driver.find_element (By.XPATH, '/html/body/main/figure/div/div[9]/div/div/div[2]/div/div/div[2]/div/button')

    descarga1.click()

    time.sleep(2)

    driver.quit()
```

```python
def gpd_ocde(): 
    url_2 = 'https://data.oecd.org/gdp/gross-domestic-product-gdp.htm'

    driver = webdriver.Chrome()

    driver.get (url_2)

    time.sleep (4)

    #data 

    clic_data = driver.find_element (By.XPATH, '//*[@id="indicator-chart"]/div[3]/div[4]/p/label/input')

    clic_data.click()

    time.sleep (2)

    #abrir desplegable

    clic_data_2 = driver.find_element (By.XPATH, '//*[@id="indicator-chart"]/div[3]/div[3]/div/a')

    clic_data_2.click()

    time.sleep (2)

    #clic removue all

    clic_data_3 = driver.find_element(By.XPATH, '//*[@id="indicator-chart"]/div[3]/div[3]/div/div/div/div[2]/p/a')

    clic_data_3.click()

    time.sleep (2)

    #Clic find country

    clic_data_4 = driver.find_element(By.XPATH, '//*[@id="indicator-chart"]/div[3]/div[3]/div/div/div/div[1]/p/label/input') #esta linea siempre será para seleccionar la barra.

    clic_data_4.click()

    time.sleep (2)

    clic_data_4.send_keys("Spain",)

    time.sleep (1)

    #Clic Spain lab

    clic_data_sp= driver.find_element (By.XPATH, '//*[@id="indicator-chart"]/div[3]/div[3]/div/div/div/div[1]/ul/li[24]/a/span[1]') #etiqueta de españa

    clic_data_sp.click() #cliquea en la etiqueta de españa.

    clic_data_4.clear() #Este ya que siempre se limpia esta celda.

    time.sleep (1)

    #francia 

    clic_data_4 = driver.find_element(By.XPATH, '//*[@id="indicator-chart"]/div[3]/div[3]/div/div/div/div[1]/p/label/input') #cliquea en la barra de búsqueda

    clic_data_4.click()

    time.sleep (2)

    clic_data_4.send_keys("France")

    #Cambiar el id path en cada pais.

    clic_data_fr= driver.find_element (By.XPATH, '//*[@id="indicator-chart"]/div[3]/div[3]/div/div/div/div[1]/ul/li[27]/a/span[1]')

    clic_data_fr.click()

    clic_data_4.clear()

    time.sleep (1)

    #UK
    clic_data_4 = driver.find_element(By.XPATH, '//*[@id="indicator-chart"]/div[3]/div[3]/div/div/div/div[1]/p/label/input') #cliquea en la barra de búsqueda

    clic_data_4.click()

    time.sleep (2)

    clic_data_4.send_keys("united Kingdom")

    clic_data_uk= driver.find_element (By.XPATH, '//*[@id="indicator-chart"]/div[3]/div[3]/div/div/div/div[1]/ul/li[28]/a/span[1]')

    clic_data_uk.click()

    clic_data_4.clear()

    time.sleep (1)

    #china

    clic_data_4 = driver.find_element(By.XPATH, '//*[@id="indicator-chart"]/div[3]/div[3]/div/div/div/div[1]/p/label/input') #cliquea en la barra de búsqueda

    clic_data_4.click()

    time.sleep (2)

    clic_data_4.send_keys("china")

    clic_data_china= driver.find_element (By.XPATH, '//*[@id="indicator-chart"]/div[3]/div[3]/div/div/div/div[1]/ul/li[15]/a/span[1]')

    clic_data_china.click()

    clic_data_4.clear()

    time.sleep (1)

    #USA

    clic_data_4 = driver.find_element(By.XPATH, '//*[@id="indicator-chart"]/div[3]/div[3]/div/div/div/div[1]/p/label/input') #cliquea en la barra de búsqueda

    clic_data_4.click()

    time.sleep (2)

    clic_data_4.send_keys("united states")

    clic_data_usa= driver.find_element (By.XPATH, '//*[@id="indicator-chart"]/div[3]/div[3]/div/div/div/div[1]/ul/li[65]/a/span[1]')

    clic_data_usa.click()

    clic_data_4.clear()

    time.sleep (1)

    #india

    clic_data_4 = driver.find_element(By.XPATH, '//*[@id="indicator-chart"]/div[3]/div[3]/div/div/div/div[1]/p/label/input') #cliquea en la barra de búsqueda

    clic_data_4.click()

    time.sleep (2)

    clic_data_4.send_keys("india")

    clic_data_usa= driver.find_element (By.XPATH, '//*[@id="indicator-chart"]/div[3]/div[3]/div/div/div/div[1]/ul/li[35]/a/span[1]')

    clic_data_usa.click()

    clic_data_4.clear()

    time.sleep (1)

    #Limpiar

    #Clic out

    clic_data_5= driver.find_element (By.XPATH, '//*[@id="indicator-chart"]/div[3]/div[3]/div/div/a/i')

    clic_data_5.click()

    time.sleep (1)

    #Mover la barra funciona!

    slider = driver.find_element(By.XPATH, '//*[@id="indicator-chart"]/div[3]/div[4]/div[1]/div')

    move = ActionChains(driver)

    width = slider.size['width']
    move.click_and_hold(slider).move_by_offset(-width*3, 0).release().perform()

    time.sleep(2)

    #clic download

    clic_data6 = driver.find_element (By.XPATH, '//*[@id="indicator-chart"]/div[1]/ul/li[3]/div/a/span')

    clic_data6.click()

    #clic descargar ya

    clic_data7 = driver.find_element (By.XPATH, '//*[@id="indicator-chart"]/div[1]/ul/li[3]/div/div/ul/li[1]/a')

    clic_data7.click()

    time.sleep(1)

    driver.quit()
```
